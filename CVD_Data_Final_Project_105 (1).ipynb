# Compiling Data

### Importing Pandas
In a code block run the two codes,
import numpy as np

import pandas as pd

This will install the pandas package onto colab with the tools manipulate your data and numpy helps pandas compute some of its math.

### Mounting Google Drive
Mounting your google drive allows consistent access to whatever files you need in pandas.To do so run the code

from google.colab import drive
drive.mount('/content/gdrive')

### Creating a Dataframe
Colab uses csv files to make dataframes for you. Using the pd.read_csv() function colab can ead the date from your CSV file into a datafram object that we can define as df.

To define the dataframe as df run the code df=dataframe

By simply copying the path of out file from Drive into the pd.read_csv() function, Pandas can now read our file.

dataframe = pd.read_csv("/content/gdrive/MyDrive/CVD_cleaned.csv")

### Filtering Data
We want to create a subset of the data we hav currently by filtering the larger dataset. We do this by runnning the code,

df[df[(name of column targeted)]== (value you are looking for)]

In my case mine looked like

df[df['Age_Category']=="80+"]

I was targeting the column regarding age category and I was looking specifically for rows that read 80+

### Naming and Exporting Data

Now that you have your subset you can define for ease of access using whatever name you choose. For example my code looks like

_80_plus_subset = df[df['Age_Category'] == "80+"]

After defining you dataset you will use the .to_csv function to convert your data into a csv file. Example:

_80_plus_subset.to_csv("_80_plus_subset.to_csv, index=False")
